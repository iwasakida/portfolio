```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
path = 'GalacticModelFDF/'
# path = 'GalacticModelFDF/n1c1o/'
# path = 'GalacticModelFDF/n1c1p/'
# path = 'GalacticModelFDF/n1c3o/'
path = 'GalacticModelFDF/n3c1o/'

files = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
        if '.dat' in file:#if '.jpg' in file:
            files.append(os.path.join(r, file))
            
print(len(files))
# files = np.random.choice(files, 10, replace=False)
files = np.sort(files)
print(files)
```

```python
# phi_new = np.arange(-49.995, 50.005, 0.01)
phi_new = np.arange(-99.995, 100.005, 0.01)

fdfs = []

for file in files:
#     print(file)
    data=np.loadtxt(file)
    phi = data[:,0]
    fdf = data[:,1]+1j*data[:,2]
    fdf_new = np.interp(phi_new, phi, fdf, 0, 0)
    fdfs.append(fdf_new)
    
fdfs = np.array(fdfs)
fdfs
```

```python
df_fdf = pd.DataFrame(fdfs, columns=phi_new)
print(df_fdf.shape)
df_fdf
```
