```python
from multiprocessing import Pool
import pandas as pd
from matplotlib import pyplot as plt
from astropy.cosmology import __path__

from astroML.datasets import fetch_sdss_spectrum
import numpy as np
import pickle
import numpy as np
```

```python
df = pd.read_csv("../RCSED/rcsed.csv")
plate = df["plate"]
mjd =df["mjd"]
fiber =df["fiberid"]
```

```python
with open('lam221000.bin', 'rb') as p:
    lam = pickle.load(p)
with open('error221000.bin', 'rb') as p:
    error = pickle.load(p)
with open('spectrum221000.bin', 'rb') as p:
    spectrum = pickle.load(p)
with open('item221000.bin', 'rb') as p:
    item = pickle.load(p)
```

```python
len(lam), len(error), len(spectrum), item.shape
```

```python
for i, j, k, l  in zip(plate[221000:250000],mjd[221000:250000],fiber[221000:250000], plate[221000:250000].index):
  
    spec = fetch_sdss_spectrum(i, j, k)

    lam.append( spec.wavelength() )
    spectrum.append( spec.spectrum ) 
    error.append( spec.error ) 

    item_append = np.array([
        l,i,j,k, spec.coeff0, spec.coeff0_restframe(), spec.coeff1, spec.large_err,
        spec.z, spec.zconf, spec.zerr,
        spec.w_max(), spec.w_min()     
    ])
    item = np.vstack((item,item_append))

    if l%1000 == 0:
        print(l)

        with open('lam{}.bin'.format(l), 'wb') as p:
            pickle.dump(lam, p)

        with open('spectrum{}.bin'.format(l), 'wb') as p:
            pickle.dump(spectrum, p)

        with open('error{}.bin'.format(l), 'wb') as p:
            pickle.dump(error, p)

        with open('item{}.bin'.format(l), 'wb') as p:
            pickle.dump(item, p)
```
